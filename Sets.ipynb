```python
Days = {"Monday", "Tuesday", "Wednesday", "Thursday", "Friday"}
print(type(Days))
print(Days)
for i in Days:
  print(i)
```

```python
Day = set(["January", "February", "March", "November"])
print(type(Day))
print(Day)
for i in Day:
  print(i)
```

```python
month = set(["December", "October", "August", "July"])
print(type(month))
month.add("June")
month.add("April")
print(month)
month.discard("July")
print(month)
month.remove("December")
print(month)
```
