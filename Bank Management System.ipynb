```python
class BankAccount:
    def __init__(self, initial_balance, account_holder):
        self._balance = initial_balance 
        self.account_holder = account_holder

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Invalid amount or Insufficient Balance!")

    def get_balance(self):
        return self._balance

class SavingsAccount(BankAccount):
    def __init__(self, initial_balance, account_holder, interest_rate):
        super().__init__(initial_balance, account_holder)
        self.interest_rate = interest_rate

    def withdraw(self, amount):
        if amount > 20000:
            print("Amount exceeds the transaction limit!")
        elif amount > self._balance:
            print("Insufficient Balance!")
        elif amount > 0:
            self._balance -= amount
            print(f"Withdrew: {amount}")

    def interest_addition(self):
        interest = self.get_balance() * self.interest_rate / 100
        self.deposit(interest)
        print("Interest is Deposited:", interest)


acc1 = BankAccount(50000, "Tom")
acc1.deposit(6700)
acc1.withdraw(2000)
print(f"Tom's Balance: {acc1.get_balance()}")

print("\n--------------")

acc2 = SavingsAccount(37000, "Vimal", 5)
acc2.deposit(4000)
acc2.withdraw(19999)
acc2.interest_addition()
print(f"Vimal's final balance: {acc2.get_balance()}")
```
